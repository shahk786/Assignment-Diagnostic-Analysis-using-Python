#### Question 2
What is the date range of the provided data sets, and which service settings reported the most appointments for a specific period?

#### Question 4
What monthly and seasonal trends are evident, based on the number of appointments for service settings, context types, and national categories?

```python
nc_service_setting = national_categories.groupby(['appointment_month', 'service_setting'], as_index=False).agg({'count_of_appointments': 'sum'})
# standardize the values to reduce skew in graph
nc_service_setting['count_of_appointments'] = nc_service_setting['count_of_appointments'] / 10**6

# to create a custom layout for the plot
fig, ax = plt.subplots()
sns.set(rc={'figure.figsize':(14, 8)})
sns.set_style('whitegrid')

# setting labels for the plot
ax.set_xlabel("Month")
ax.set_ylabel("Number of Appointments (in millions)")
ax.set_title("Appointments per Month per Service Setting")

sns.lineplot(data=nc_service_setting, x='appointment_month', y='count_of_appointments', hue='service_setting')
plt.show()
```

```python
nc_context_types = national_categories.groupby(['appointment_month', 'context_type'], as_index=False).agg({'count_of_appointments': 'sum'})
# standardize the values to reduce skew in graph
nc_context_types['count_of_appointments'] = nc_context_types['count_of_appointments'] / 10**6

# to create a custom layout for the plot
fig, ax = plt.subplots()
sns.set(rc={'figure.figsize':(10, 6)})
sns.set_style('whitegrid')

# setting labels for the plot
ax.set_xlabel("Month")
ax.set_ylabel("Number of Appointments (in millions)")
ax.set_title("Appointments per Month per Context Type")

sns.lineplot(data=nc_context_types, x='appointment_month', y='count_of_appointments', hue='context_type')
plt.show()
```

```python
nc_national_category = national_categories.groupby(['appointment_month', 'national_category'], as_index=False).agg({'count_of_appointments': 'sum'})
# standardize the values to reduce skew in graph
nc_national_category['count_of_appointments'] = nc_national_category['count_of_appointments'] / 10**6

# to create a custom layout for the plot
fig, ax = plt.subplots()
sns.set(rc={'figure.figsize':(10, 6)})
sns.set_style('whitegrid')

# setting labels for the plot
ax.set_xlabel("Month")
ax.set_ylabel("Number of Appointments (in millions)")
ax.set_title("Appointments per Month per National Category")

sns.lineplot(data=nc_national_category, x='appointment_month', y='count_of_appointments', hue='national_category')
plt.legend(bbox_to_anchor = (2.30, 0.6), loc='center right')
plt.tight_layout()
plt.show()
```

#### Question 5
What are the top trending hashtags (#) on Twitter related to healthcare in the UK?

```python
# filter out all records/tweets that have a hashtag i.e hashtag field is not null
tweets_with_nonnull_hashtags = tweets[tweets.tweet_entities_hashtags.notnull()]
# convert all the tweets into a string and then list (this is done to account for tweets with multiple hashtags)
tweets_hashtags_string = ', '.join(tweets_with_nonnull_hashtags.tweet_entities_hashtags)
tweets_hashtags_list = tweets_hashtags_string.split(', ')

# convert to pandas series to make use of easy count functionality
tweets_hashtags_series = pd.Series(tweets_hashtags_list)

tweets_most_df = pd.DataFrame(tweets_hashtags_series.value_counts().head(10)).reset_index()
tweets_most_df.columns = ['hashtag', 'count_hashtags']
```

```python
sns.barplot(data=tweets_most_df, x='hashtag', y='count_hashtags')
plt.show()
```

```python
hcp_per_icb_per_month = appointments_regional.groupby(['icb_ons_code', 'hcp_type', 'appointment_month'], as_index=False).agg({'count_of_appointments': 'sum', 'appointment_status': 'count'})
hcp_per_icb_per_month.columns = ['icb_ons_code', 'hcp_type', 'appointment_month', 'number_of_appointments', 'number_of_healthcare_professionals_slots']
hcp_per_icb_per_month['capacity'] = hcp_per_icb_per_month['number_of_appointments'] / hcp_per_icb_per_month['number_of_healthcare_professionals_slots']

hcp_per_icb_per_month
```

```python
hcp_capacity = national_categories[['icb_ons_code', 'sub_icb_location_name']].merge(hcp_per_icb_per_month, on='icb_ons_code')
```
